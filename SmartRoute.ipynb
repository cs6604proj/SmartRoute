```python
import urllib2
import urllib
import json
import time

PLACE_BASE_URL = "https://maps.googleapis.com/maps/api/place/nearbysearch/json?"
DIRECTION_BASE_URL = "https://maps.googleapis.com/maps/api/directions/json?"
API_KEY = "AIzaSyA0jerx_Cm8VpwkVQqNtu2Za7NDU05JVzk"


print nearby_search

def get_pois(location, keyword):
    nearby_param = {"key": API_KEY, 
                "location":location,
                "radius": 3200,
                "sensor": "false",
                "rankby": "prominence",
                "keyword": keyword}
    nearby_search = PLACE_BASE_URL + urllib.urlencode(nearby_param)
    response = json.load(urllib2.urlopen(nearby_search))
    if response["status"] != "OK":
        return None
    pois = []
    for result in response["results"]:
        poi = {"geometry": result["geometry"],
               "name": result["name"],
               "price_level": result.get("price_level", None),
               "rating": result.get("rating", None),
               "vicinity": result["vicinity"]}
        pois.append(poi)
    return pois

locations = ['38.9015775,-77.2087283', '38.8859531663,-77.1568830199', 
             '38.8821828738,-77.113168835', '38.8959790962,-77.0709086853', 
             '38.950928,-77.38362', '38.962705,-77.433688']

way_points = []
keywords = {"startbucks", "supermarket"}
for location in locations:
    information = {"location": location}
    for keyword in keywords:
        
        information[keyword] = get_pois(location, keyword)
    way_points.append(information)   
print way_points        
    
    
```
